```python
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from transformers import TrOCRProcessor, VisionEncoderDecoderModel

# Load TrOCR pre-trained model
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-handwritten")

# Function to process and recognize handwritten text
def recognize_handwritten_text():
    print("📤 Upload a handwritten text image...")

    uploaded = files.upload()

    for filename in uploaded.keys():
        # Load image in grayscale
        img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

        # Convert grayscale to RGB (TrOCR requires RGB)
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

        # Resize image for model input
        img = cv2.resize(img, (800, 400))

        # Show the processed image
        plt.imshow(img, cmap='gray')
        plt.title("Processed Handwritten Text")
        plt.show()

        # Convert to tensor format for TrOCR
        pixel_values = processor(images=img, return_tensors="pt").pixel_values

        # Generate prediction
        with torch.no_grad():
            predicted_ids = model.generate(pixel_values)

        # Convert IDs to text
        recognized_text = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
        print(f"📝 Recognized Text: {recognized_text}")

# Run the function
recognize_handwritten_text()

```

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
from google.colab import files

# Upload and recognize handwritten text
def recognize_handwritten_text():
    print("📤 Upload a handwritten text image...")

    uploaded = files.upload()

    for filename in uploaded.keys():
        # Load image in grayscale
        img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

        # Resize for better accuracy
        img = cv2.resize(img, (800, 400))

        # Apply thresholding to remove noise
        _, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Show the processed image
        plt.imshow(img, cmap='gray')
        plt.title("Processed Handwritten Text")
        plt.show()

        # Use Tesseract OCR to recognize text
        text = pytesseract.image_to_string(img, config='--psm 6')  # psm 6 is best for blocks of text
        print(f"📝 Recognized Text:\n{text}")

# Run the function
recognize_handwritten_text()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

def preprocess_and_predict():
    print("📤 Upload a handwritten digit image...")

    uploaded = files.upload()

    for filename in uploaded.keys():
        # Load image in grayscale
        img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

        # Resize to 28x28 pixels
        img = cv2.resize(img, (28, 28))

        # Apply Gaussian Blur (reduces noise)
        img = cv2.GaussianBlur(img, (5,5), 0)

        # Adaptive Thresholding to improve contrast
        img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                    cv2.THRESH_BINARY_INV, 11, 2)

        # Normalize the image
        img = img / 255.0

        # Reshape for model input
        img = img.reshape(1, 28, 28, 1)

        # Show the processed image
        plt.imshow(img.squeeze(), cmap='gray')
        plt.title("Enhanced Processed Image")
        plt.show()

        # Predict the digit
        prediction = model.predict(img)
        predicted_digit = np.argmax(prediction)
        print(f"📝 Recognized Digit: {predicted_digit}")

# Run the function again
preprocess_and_predict()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize pixel values
x_train, x_test = x_train / 255.0, x_test / 255.0

# Reshape to match CNN input shape
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Improved CNN Model
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),  # Additional convolution layer
    layers.Flatten(),

    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),  # Dropout to prevent overfitting
    layers.Dense(10, activation='softmax')
])

# Compile Model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train Model with More Epochs
history = model.fit(x_train, y_train, epochs=15, validation_data=(x_test, y_test))

# Evaluate Model
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"🔍 Model Test Accuracy: {test_acc:.4f}")

# Plot Training History
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
